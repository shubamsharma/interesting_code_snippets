# Count No of "True" in dataframe row

```python
import numpy as np
import pandas as pd


data = {'a': [True,True,True],'b': [True,False,True],'c': [True,True,True]}
df = pd.DataFrame(data)
df['count'] = df[df == True].sum(axis = 1)
df['count'] = np.where(df['count']>2,1,0)
```

### Compute agg percentage and merge interval based on consecutive True present in value column


```python
import pandas as pd
data1 = pd.Interval(0, 1, closed='right')
data2 = pd.Interval(1, 2, closed='right')
data3 = pd.Interval(2, 3, closed='right')
data4 = pd.Interval(3, 4, closed='right')
data5 = pd.Interval(4, 5, closed='right')
data6 = pd.Interval(5, 6, closed='right')
data7 = pd.Interval(6, 7, closed='right')
data8 = pd.Interval(7, 8, closed='right')
data9 = pd.Interval(8, 9, closed='right')
data10 = pd.Interval(9, 10, closed='right')

data = {'interval_data' : [data1,data2,data3,data4,data5,data6,data7,data8,data9,data10],
        'percent' : [.01,.02,1.03,.10,.05,1.07,20,30,.5,.5],
        "value" : [True,True,False,True,True,False,False,False,True,True]}

data1 = pd.DataFrame(data)

data1['lower'] = data1.interval_data.apply(lambda x : x.left)
data1['upper'] = data1.interval_data.apply(lambda x : x.right)

data1['value_u_shift'] = (data1['value'].shift(-1)).replace(np.nan, False).astype(bool)
data1['value_l_shift'] = data1['value'].shift(1).replace(np.nan, False).astype(bool)

data1['count'] = data1[['value','value_u_shift','value_l_shift']][data1 == True].sum(axis = 1)
data1['count'] = np.where(data1['count']>=2,1,0)
data1['count'] = np.where(data1['value']==False,0,data1['count'])

counter = 1
def funct(var):
    global counter
    if var == 1:
        return(counter) 
    else :
        counter = counter + 1
        return(var)
data1['n_count'] = data1['count'].apply(funct)

print("Output")
data1[data1['n_count']!=0].groupby(['n_count']).agg(agg_percent=("percent",sum),
                                                   lower_n = ("lower", lambda a : a.head(1)),
                                                   upper_n = ("upper", lambda a : a.tail(1))).reset_index()
```
